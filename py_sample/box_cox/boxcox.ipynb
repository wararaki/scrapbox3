```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
r = stats.skewnorm.rvs(5.5, loc=0.5, scale=0.5, size=1000)
```

```python
plt.hist(r, density=True, histtype='stepfilled', alpha=0.5, bins=100)
plt.show()
```

```python
y = stats.boxcox(r, lmbda=0.2)
```

```python
plt.hist(y, density=True, histtype='stepfilled', alpha=0.5, bins=100)
plt.show()
```
