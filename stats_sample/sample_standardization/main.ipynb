```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris.feature_names
```

```python
X = iris.data[:, 2:]
X.shape
```

```python
mu = X.mean(axis=0)
mu
```

```python
Sigma = np.cov(X, rowvar=0)
Sigma
```

```python
sigma = np.diag(Sigma)
sigma
```

```python
z = (X-mu)/sigma
z.shape
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
plt.scatter(z[:, 0], z[:, 1])
plt.show()
```
