```python
import glob
from typing import Dict

import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import tqdm

%matplotlib inline
```

```python
class Loader:
    def __init__(self, pathname: str):
        self._pathname = pathname

    def _convert(self, line: str) -> Dict[str, float]:
        items = line.split()
        features = {
            f'feature_{int(feature[0]):03d}': float(feature[1]) for feature in map(
                lambda x: x.split(':'), items[2:]
            )
        }
        features['relevance'] = int(items[0])
        features['qid'] = int(items[1].split(':')[1])
        return features

    def __iter__(self) -> Dict[str, float]:
        for filename in glob.glob(self._pathname):
            with open(filename) as f:
                for line in f:
                    yield self._convert(line)
```

```python
loader = Loader('data/Fold1/test.*')
```

```python
data = []
for feature in loader:
    data.append(feature)
len(data)
```
