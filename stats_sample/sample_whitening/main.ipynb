```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris.feature_names
```

```python
X = iris.data[:, 2:]
X.shape
```

```python
Sigma = np.cov(X, rowvar=0)
Sigma
```

```python
_, S = np.linalg.eig(Sigma)
S
```

```python
mu = X.mean(axis=0)
mu
```

```python
lambda_ = np.dot(np.dot(S.T, Sigma), S)
lambda_
```

```python
lambda__ = np.linalg.inv(np.sqrt(lambda_))
lambda__
```

```python
u = np.dot(np.dot(lambda__, S.T), (X-mu).T).T
u.shape
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
plt.scatter(u[:, 0], u[:, 1])
plt.show()
```
