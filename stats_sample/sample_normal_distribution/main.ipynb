```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu = 0.0
sigma = 0.1
```

```python
s = np.random.normal(mu, sigma, 1000)
```

```python
s[:10]
```

```python
count, bins, ignored = plt.hist(s, 30, density=True)
```

```python
def normal(x: np.ndarray, mu: float, sigma: float) -> np.ndarray:
    return 1/(np.sqrt(2*np.pi)*sigma) * np.exp(-np.power(x-mu, 2)/(2*np.power(sigma,2)))
```

```python
print(bins)
y = normal(bins, mu, sigma)
print(y)
```

```python
plt.plot(bins, y)
```

```python
x_2 = np.linspace(-1, 1, num=10000)
x_2
```

```python
y_2 = normal(x_2, mu, sigma)
y_2
```

```python
plt.plot(x_2, y_2)
plt.show()
```

```python
y_3 = normal(x_2, 0.5, sigma)
y_3
```

```python
plt.plot(x_2, y_3)
plt.show()
```

```python
y_4 = normal(x_2, mu, 0.5)
y_4
```

```python
plt.plot(x_2, y_4)
plt.show()
```

iris データセットの最初の feature の分布を見る。

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
x_iris = iris.data[:, 0]
```

```python
mu_iris = x_iris.mean()
sigma_iris = np.sqrt(x_iris.var())
print(mu_iris, sigma_iris)
```

```python
sample = np.linspace(3, 9, num=10000)
y_iris = normal(sample, mu_iris, sigma_iris)   
```

```python
plt.plot(sample, y_iris)
plt.show()
```
